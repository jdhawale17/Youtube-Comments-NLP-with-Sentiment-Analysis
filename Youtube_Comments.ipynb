```python
import pandas as pd
import googleapiclient.discovery

api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = "AIzaSyAAafNd3TnZeZTlUibeuAVYnXNFZkoKn_w"

youtube = googleapiclient.discovery.build(api_service_name, api_version, developerKey = DEVELOPER_KEY)

request = youtube.commentThreads().list(
    part = "snippet",
    videoId = "FgakZw6K1QQ",
    maxResults = 100)
response = request.execute()

comments = []

for item in response['items']:
    comment = item['snippet']['topLevelComment']['snippet']
    comments.append([
        comment['authorDisplayName'],
        comment['publishedAt'],
        comment['updatedAt'],
        comment['likeCount'],
        comment['textDisplay']
    ])


df = pd.DataFrame(comments, columns=['Author', 'published_at', 'updated_at','like_count','text'])
print(df)
```

```python
df.size
df.head(10)
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

df['sentiment_scores'] = df['text'].apply(lambda text: sia.polarity_scores(text))
df['compound_score'] = df['sentiment_scores'].apply(lambda score_dict: score_dict['compound'])
print(df)
```

```python
# prompt: in df, split sentiment scores into 3 separate columns, based on 'neg', 'neu', and 'pos'

df['neg'] = df['sentiment_scores'].apply(lambda score_dict: score_dict['neg'])
df['neu'] = df['sentiment_scores'].apply(lambda score_dict: score_dict['neu'])
df['pos'] = df['sentiment_scores'].apply(lambda score_dict: score_dict['pos'])
df
```

```python
# prompt: Can you order df by 'neg', largest to smallest

df = df.sort_values(by='neg', ascending=False)
print(df)
```
